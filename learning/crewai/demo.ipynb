> **YAAF** (Yet another agent framework)
CrewAI is a nice way to orchestrate agentic workloads, built on top of Langchain. 



```python
import os
import pprint as pp

from dotenv import load_dotenv, find_dotenv
from crewai import Agent, Task, Crew, Process

from langchain.llms import Ollama
from langchain.tools import DuckDuckGoSearchRun
from langchain.agents import AgentType, initialize_agent, load_tools

load_dotenv(find_dotenv(filename="../../.env"))

search_tool = DuckDuckGoSearchRun()
ollama = Ollama(model="openhermes")
```

```python
# Define the agents
researcher_agent = Agent(
    role="Researcher",
    goal="Research new insights related to AI",
    backstory="you are an AI assistant researcher",
    verbose=True,
    allow_delegation=False,
    llm = ollama,
    tools=[search_tool]
)

writer_agent = Agent(
    role="Writer",
    goal="Write blogposts in a natural and human way about AI topics",
    backstory="You write amazing blogposts about AI for a living.",
    verbose=True,
    allow_delegation=False,
    llm = ollama,
    tools=[search_tool]
)
```

```python
# Create the different takss
task1 = Task(
    description="investigate latest AI trends", agent=researcher_agent
)
task2 = Task(
    description="Write a compelling blogpost about AI trends", agent=writer_agent
)
```

```python
# Need to make the crew (similar to a groupchat)
crew = Crew(
    agents=[researcher_agent, writer_agent],
    tasks=[task1, task2],
    verbose=2,
    process=Process.sequential
)
```

```python
# Start the process
result = crew.kickoff()
result
```

```python
pp.pprint(result)
```

### Using crewai to research ArXiv for recent papers.

```python

# Initialize ArXiv tool
tools = load_tools(
    ["arxiv"],
)

# Define the agents
research_paper_finder = Agent(
    role='Research Paper Finder',
    goal='Find top 10 recent papers in 2024 on LLM, generative AI, and agent-related topics',
    backstory='As a researcher, you are constantly seeking the latest advancements in AI.',
    tools=tools,
    verbose=True,
    llm=ollama
)

ai_blog_writer = Agent(
    role='AI Blog Writer',
    goal='Write a blog post about recent research findings in LLM, generative AI, and agent-related topics',
    backstory='You are a writer specializing in AI technology and its latest trends.',
    verbose=True,
    llm=ollama
)

# Define the tasks
task1 = Task(
    description='Find recent top research papers on LLM, generative AI, and agent-related topics using ArXiv.',
    agent=research_paper_finder
)

task2 = Task(
    description='Analyze the papers found by the Research Paper Finder and write an informative blog post summarizing the key findings and their implications.',
    agent=ai_blog_writer
)

# Instantiate the crew with a sequential process
crew = Crew(
    agents=[research_paper_finder, ai_blog_writer],
    tasks=[task1, task2],
    process=Process.sequential,
    verbose=True
)

# Start the crew
result = crew.kickoff()
pp.pprint(result)
```
